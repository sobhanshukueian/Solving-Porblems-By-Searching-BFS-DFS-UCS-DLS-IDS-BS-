# DFS (Depth-first search)
Depth-first search (DFS) is an algorithm for traversing or searching tree or graph data structures. The algorithm starts at the root node (selecting some arbitrary node as the root node in the case of a graph) and explores as far as possible along each branch before backtracking. 

Depth-First-Search.gif

[More About Algorithm](https://en.wikipedia.org/wiki/Depth-first_search)

```python
class Graph:
    def __init__(self):
        self.graph = {}
        self.result = []
        self.visited = []

    def add_edge(self, u, v):
        if u not in self.graph.keys():
            self.graph[u] = []
        self.graph[u].append(v)

    def get_graph(self):
        return self.graph

    def dfs(self, root, goal):
        if root not in self.visited:
            self.result.append(str(root))
            self.visited.append(root)
            if root in  self.graph.keys():
                for neighbour in self.graph[root]:
                    if goal in self.visited:
                        break
                    else:
                        self.dfs(neighbour, goal)
        return self.result

g = Graph()
g.add_edge(1 ,2)
g.add_edge(1 ,4)
g.add_edge(1 ,3)
g.add_edge(2 ,4)
g.add_edge(3 ,5)
print(g.get_graph())
" -> ".join(g.dfs(1, 5))
```
